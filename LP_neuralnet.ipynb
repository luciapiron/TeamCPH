```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_percentage_error

import tensorflow as tf
from tensorflow.keras import layers, models, callbacks


# -----------------------------
# CONFIG
# -----------------------------

DATA_PATH = "/workspaces/TeamCPH/data/combined_df_corrected.csv"

TARGET_COL = "Revenue"
DATE_COL = "Date"

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

EPOCHS = 80
BATCH_SIZE = 64
LEARNING_RATE = 1e-3

# Time split proportions (time-based)
TRAIN_FRAC = 0.70
VAL_FRAC = 0.15
TEST_FRAC = 0.15


# -----------------------------
# 1) LOAD DATA
# -----------------------------
df = pd.read_csv(DATA_PATH)
df[DATE_COL] = pd.to_datetime(df[DATE_COL])
df = df.sort_values(DATE_COL).reset_index(drop=True)

print("Loaded data shape:", df.shape)
print(df.head())


# -----------------------------
# 2) PRODUCT GROUP LABEL
# -----------------------------
# Dataset contains Product_2..Product_6 booleans; if none are True -> Product_1
product_cols = [c for c in df.columns if c.startswith("Product_")]

def infer_product_group(row):
    for c in product_cols:
        if bool(row[c]):
            return c.replace("Product_", "Product_")
    return "Product_1"

df["Product_Group"] = df.apply(infer_product_group, axis=1)
print("\nProduct group distribution:")
print(df["Product_Group"].value_counts())


# -----------------------------
# 3) BASIC CLEANING / FEATURE PREP
# -----------------------------
# Convert boolean columns to 0/1
bool_cols = df.select_dtypes(include=["bool"]).columns.tolist()
for c in bool_cols:
    df[c] = df[c].astype(int)

# Handle missing values in lag features (e.g., lag_1 is NaN for first row)
# Here: forward fill, then fallback to 0 if still missing.
df = df.fillna(method="ffill").fillna(0)

# Drop Date and Product_Group from features
feature_cols = [c for c in df.columns if c not in [TARGET_COL, DATE_COL, "Product_Group"]]

X = df[feature_cols].copy()
y = df[TARGET_COL].copy()


# -----------------------------
# 4) TIME-BASED SPLIT
# -----------------------------
n = len(df)
train_end = int(n * TRAIN_FRAC)
val_end = int(n * (TRAIN_FRAC + VAL_FRAC))

X_train, y_train = X.iloc[:train_end], y.iloc[:train_end]
X_val, y_val = X.iloc[train_end:val_end], y.iloc[train_end:val_end]
X_test, y_test = X.iloc[val_end:], y.iloc[val_end:]

val_groups = df["Product_Group"].iloc[train_end:val_end].reset_index(drop=True)

print("\nSplit sizes:")
print("Train:", X_train.shape, y_train.shape)
print("Val:  ", X_val.shape, y_val.shape)
print("Test: ", X_test.shape, y_test.shape)


# -----------------------------
# 5) SCALE NUMERIC FEATURES
# -----------------------------
# We scale all feature columns. If you want to scale only some columns, customize here.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)


# -----------------------------
# 6) DEFINE NEURAL NETWORK
# -----------------------------
def build_model(input_dim: int):
    model = models.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.2),
        layers.Dense(64, activation="relu"),
        layers.Dropout(0.2),
        layers.Dense(32, activation="relu"),
        layers.Dense(1)  # regression output
    ])
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
        loss="mse",
        metrics=[tf.keras.metrics.MeanAbsolutePercentageError(name="mape")]
    )
    return model

model = build_model(X_train_scaled.shape[1])
model.summary()


# -----------------------------
# 7) TRAIN
# -----------------------------
early_stop = callbacks.EarlyStopping(
    monitor="val_loss",
    patience=12,
    restore_best_weights=True
)

history = model.fit(
    X_train_scaled, y_train,
    validation_data=(X_val_scaled, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=[early_stop],
    verbose=1
)


# -----------------------------
# 8) PLOT LOSS CURVES
# -----------------------------
plt.figure()
plt.plot(history.history["loss"], label="Train Loss (MSE)")
plt.plot(history.history["val_loss"], label="Val Loss (MSE)")
plt.title("Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.legend()
plt.grid(True)
plt.show()


# -----------------------------
# 9) VALIDATION METRICS: OVERALL MAPE + PER PRODUCT GROUP
# -----------------------------
val_preds = model.predict(X_val_scaled).reshape(-1)

overall_val_mape = mean_absolute_percentage_error(y_val, val_preds) * 100
print(f"\nOverall Validation MAPE: {overall_val_mape:.3f}%")

# Build a results dataframe for group evaluation
val_results = pd.DataFrame({
    "y_true": y_val.reset_index(drop=True),
    "y_pred": val_preds,
    "Product_Group": val_groups
})

group_mapes = (
    val_results
    .groupby("Product_Group")
    .apply(lambda g: mean_absolute_percentage_error(g["y_true"], g["y_pred"]) * 100)
    .sort_index()
)

print("\nValidation MAPE by Product Group:")
print(group_mapes)


# -----------------------------
# 10) OPTIONAL: TEST EVALUATION
# -----------------------------
test_preds = model.predict(X_test_scaled).reshape(-1)
overall_test_mape = mean_absolute_percentage_error(y_test, test_preds) * 100
print(f"\nOverall Test MAPE: {overall_test_mape:.3f}%")


# -----------------------------
# 11) OPTIONAL: SAVE MODEL + SCALER
# -----------------------------
# model.save("revenue_forecast_nn.keras")
# import joblib
# joblib.dump(scaler, "feature_scaler.pkl")

print("\nDone.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_percentage_error, mean_absolute_error, mean_squared_error


# -----------------------------
# Helper metrics
# -----------------------------
def safe_mape(y_true, y_pred, eps=1e-6):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    denom = np.maximum(np.abs(y_true), eps)
    return np.mean(np.abs((y_true - y_pred) / denom)) * 100


# -----------------------------
# 0) Convert to aligned structures
# -----------------------------
y_val_arr = np.array(y_val).reshape(-1)
val_preds_arr = np.array(val_preds).reshape(-1)

# If val_groups is a Series, align index
val_groups_arr = np.array(val_groups).reshape(-1)

residuals = y_val_arr - val_preds_arr

# Exclude zeros for standard MAPE (your teacher requirement)
nonzero_mask = y_val_arr != 0


# -----------------------------
# 1) TRAINING CURVES (Loss)
# -----------------------------
plt.figure()
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()


# -----------------------------
# 2) ACTUAL vs PREDICTED SCATTER
# -----------------------------
plt.figure()
plt.scatter(y_val_arr, val_preds_arr, alpha=0.4)
max_val = max(y_val_arr.max(), val_preds_arr.max())
min_val = min(y_val_arr.min(), val_preds_arr.min())
plt.plot([min_val, max_val], [min_val, max_val])  # diagonal
plt.title("Actual vs Predicted (Validation)")
plt.xlabel("Actual Revenue")
plt.ylabel("Predicted Revenue")
plt.grid(True)
plt.show()


# -----------------------------
# 3) RESIDUAL HISTOGRAM
# -----------------------------
plt.figure()
plt.hist(residuals, bins=50)
plt.title("Residual Distribution (Validation)")
plt.xlabel("Residual = Actual - Predicted")
plt.ylabel("Count")
plt.grid(True)
plt.show()


# -----------------------------
# 4) RESIDUALS vs PREDICTED (BIAS CHECK)
# -----------------------------
plt.figure()
plt.scatter(val_preds_arr, residuals, alpha=0.4)
plt.axhline(0)
plt.title("Residuals vs Predicted (Validation)")
plt.xlabel("Predicted Revenue")
plt.ylabel("Residual (Actual - Predicted)")
plt.grid(True)
plt.show()


# -----------------------------
# 5) ACTUAL vs PREDICTED OVER TIME
# -----------------------------
# If you have the original validation dates, use them:
# In my earlier script the val split is based on indices; we reconstruct the date slice.
# Replace these indices if your split differs.

# If you saved train_end and val_end earlier, use them.
# Otherwise infer by length:
val_len = len(y_val_arr)

# Attempt to locate the validation set in df by matching y_val values
# Better: use slicing with train_end and val_end in your main script.
# Here is the recommended approach:
try:
    val_dates = df["Date"].iloc[-(len(df) - len(y_val_arr) - len(y_test)) : -(len(y_test))].reset_index(drop=True)
except Exception:
    # fallback: just create a numeric x-axis
    val_dates = pd.Series(range(val_len), name="ValidationIndex")

plt.figure()
plt.plot(val_dates, y_val_arr, label="Actual")
plt.plot(val_dates, val_preds_arr, label="Predicted")
plt.title("Actual vs Predicted Over Validation Time")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.legend()
plt.grid(True)
plt.show()


# -----------------------------
# 6) MAPE BY PRODUCT GROUP (BAR CHART)
# -----------------------------
val_results = pd.DataFrame({
    "y_true": y_val_arr,
    "y_pred": val_preds_arr,
    "Product_Group": val_groups_arr
})

# Standard MAPE excluding zeros
def group_mape(g):
    g = g[g["y_true"] != 0]
    if len(g) == 0:
        return np.nan
    return mean_absolute_percentage_error(g["y_true"], g["y_pred"]) * 100

group_mape_scores = val_results.groupby("Product_Group").apply(group_mape).sort_index()

plt.figure()
plt.bar(group_mape_scores.index, group_mape_scores.values)
plt.title("Validation MAPE by Product Group (excluding y=0)")
plt.xlabel("Product Group")
plt.ylabel("MAPE (%)")
plt.grid(True, axis="y")
plt.show()

print("\nMAPE by Product Group:")
print(group_mape_scores)


# -----------------------------
# 7) ABSOLUTE PERCENT ERROR DISTRIBUTION (BOXPLOT BY GROUP)
# -----------------------------
# APE = |(y - y_hat)/y|
val_results_nonzero = val_results[val_results["y_true"] != 0].copy()
val_results_nonzero["APE_%"] = (np.abs(val_results_nonzero["y_true"] - val_results_nonzero["y_pred"])
                               / np.abs(val_results_nonzero["y_true"])) * 100

groups = sorted(val_results_nonzero["Product_Group"].unique())
data = [val_results_nonzero[val_results_nonzero["Product_Group"] == g]["APE_%"].values for g in groups]

plt.figure()
plt.boxplot(data, labels=groups, showfliers=False)
plt.title("Absolute Percentage Error Distribution by Product Group")
plt.xlabel("Product Group")
plt.ylabel("Absolute Percentage Error (%)")
plt.grid(True, axis="y")
plt.show()


# -----------------------------
# 8) QUICK SUMMARY METRICS
# -----------------------------
mae = mean_absolute_error(y_val_arr, val_preds_arr)
rmse = np.sqrt(mean_squared_error(y_val_arr, val_preds_arr))

# Standard MAPE excluding y=0
if nonzero_mask.sum() > 0:
    mape_excl_zero = mean_absolute_percentage_error(y_val_arr[nonzero_mask], val_preds_arr[nonzero_mask]) * 100
else:
    mape_excl_zero = np.nan

print("\nValidation Summary Metrics:")
print(f"MAE  = {mae:,.3f}")
print(f"RMSE = {rmse:,.3f}")
print(f"MAPE (excluding y=0) = {mape_excl_zero:,.3f}%")
print(f"Safe-MAPE (epsilon)  = {safe_mape(y_val_arr, val_preds_arr):,.3f}%")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_percentage_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, BatchNormalization, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping


# -----------------------------
# CONFIG
# -----------------------------
DATA_PATH = "/workspaces/TeamCPH/data/combined_df_corrected.csv"
TARGET_COL = "Revenue"
DATE_COL = "Date"

EPOCHS = 80
BATCH_SIZE = 64
LEARNING_RATE = 0.001

TRAIN_FRAC = 0.70
VAL_FRAC = 0.15
TEST_FRAC = 0.15

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)


# -----------------------------
# 1) LOAD DATA
# -----------------------------
df = pd.read_csv(DATA_PATH)
df[DATE_COL] = pd.to_datetime(df[DATE_COL])
df = df.sort_values(DATE_COL).reset_index(drop=True)

print("Loaded data shape:", df.shape)
print(df.head())


# -----------------------------
# 2) PRODUCT GROUP LABEL
# -----------------------------
product_cols = [c for c in df.columns if c.startswith("Product_")]

# If none of Product_2..Product_6 is True -> Product_1
def infer_product_group(row):
    for c in product_cols:
        if bool(row[c]):
            return c
    return "Product_1"

df["Product_Group"] = df.apply(infer_product_group, axis=1)

print("\nProduct group distribution:")
print(df["Product_Group"].value_counts())


# -----------------------------
# 3) CLEANING / FEATURE SETUP
# -----------------------------
# Convert boolean columns to int
bool_cols = df.select_dtypes(include=["bool"]).columns.tolist()
for c in bool_cols:
    df[c] = df[c].astype(int)

# Fill missing values (common for lag features)
df = df.fillna(method="ffill").fillna(0)

# Features exclude target, date, product group label
feature_cols = [c for c in df.columns if c not in [TARGET_COL, DATE_COL, "Product_Group"]]

X = df[feature_cols].copy()
y = df[TARGET_COL].copy()


# -----------------------------
# 4) TIME-BASED TRAIN / VAL / TEST SPLIT
# -----------------------------
n = len(df)
train_end = int(n * TRAIN_FRAC)
val_end = int(n * (TRAIN_FRAC + VAL_FRAC))

X_train, y_train = X.iloc[:train_end], y.iloc[:train_end]
X_val, y_val = X.iloc[train_end:val_end], y.iloc[train_end:val_end]
X_test, y_test = X.iloc[val_end:], y.iloc[val_end:]

val_groups = df["Product_Group"].iloc[train_end:val_end].reset_index(drop=True)

print("\nSplit sizes:")
print("Train:", X_train.shape, y_train.shape)
print("Val:  ", X_val.shape, y_val.shape)
print("Test: ", X_test.shape, y_test.shape)


# -----------------------------
# 5) SCALE FEATURES
# -----------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)


# -----------------------------
# 6) BUILD MODEL (matching notebook style)
# -----------------------------
model = Sequential()
model.add(InputLayer(input_shape=(X_train_scaled.shape[1],)))
model.add(Dense(128, activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(64, activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.2))

model.add(Dense(32, activation="relu"))
model.add(Dense(1))  # regression output

model.compile(loss="mse", optimizer=Adam(learning_rate=LEARNING_RATE))

model.summary()


# -----------------------------
# 7) TRAIN MODEL
# -----------------------------
early_stop = EarlyStopping(
    monitor="val_loss",
    patience=12,
    restore_best_weights=True
)

history = model.fit(
    X_train_scaled, y_train,
    validation_data=(X_val_scaled, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=[early_stop],
    verbose=1
)


# -----------------------------
# 8) PLOT LOSS CURVES
# -----------------------------
plt.figure(figsize=(10, 5))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Training and Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.legend()
plt.grid(True)
plt.show()


# -----------------------------
# 9) PREDICTIONS
# -----------------------------
train_preds = model.predict(X_train_scaled).reshape(-1)
val_preds = model.predict(X_val_scaled).reshape(-1)
test_preds = model.predict(X_test_scaled).reshape(-1)


# -----------------------------
# 10) MAPE FUNCTIONS (teacher requirement)
# -----------------------------
# IMPORTANT: Standard MAPE is undefined for y=0.
# We compute teacher MAPE only on rows where y != 0.

def mape_excluding_zeros(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    mask = y_true != 0
    if mask.sum() == 0:
        return np.nan
    return mean_absolute_percentage_error(y_true[mask], y_pred[mask]) * 100


train_mape = mape_excluding_zeros(y_train, train_preds)
val_mape = mape_excluding_zeros(y_val, val_preds)
test_mape = mape_excluding_zeros(y_test, test_preds)

print("\nMAPE (excluding y=0):")
print(f"Training MAPE:   {train_mape:.3f}%")
print(f"Validation MAPE: {val_mape:.3f}%")
print(f"Test MAPE:       {test_mape:.3f}%")


# -----------------------------
# 11) MAPE BY PRODUCT GROUP (VALIDATION)
# -----------------------------
val_results = pd.DataFrame({
    "y_true": y_val.reset_index(drop=True),
    "y_pred": val_preds,
    "Product_Group": val_groups
})

def group_mape(g):
    g = g[g["y_true"] != 0]
    if len(g) == 0:
        return np.nan
    return mean_absolute_percentage_error(g["y_true"], g["y_pred"]) * 100

group_mape_scores = val_results.groupby("Product_Group").apply(group_mape).sort_index()

print("\nValidation MAPE by Product Group (excluding y=0):")
print(group_mape_scores)


# -----------------------------
# 12) BAR CHART: MAPE BY PRODUCT GROUP
# -----------------------------
plt.figure(figsize=(8, 5))
plt.bar(group_mape_scores.index, group_mape_scores.values)
plt.title("Validation MAPE by Product Group (excluding y=0)")
plt.xlabel("Product Group")
plt.ylabel("MAPE (%)")
plt.grid(True, axis="y")
plt.show()


# -----------------------------
# 13) OPTIONAL: ACTUAL vs PREDICTED (VALIDATION)
# -----------------------------
plt.figure(figsize=(10, 5))
plt.plot(y_val.reset_index(drop=True).values, label="Actual", linewidth=2)
plt.plot(val_preds, label="Predicted", linewidth=2)
plt.title("Validation: Actual vs Predicted Revenue")
plt.xlabel("Validation Sample Index")
plt.ylabel("Revenue")
plt.legend()
plt.grid(True)
plt.show()

print("\nDone.")
```
